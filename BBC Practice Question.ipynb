2.	Write a Pandas program to select first 2 rows, 2 columns and specific two columns from World alcohol consumption dataset

```python
alcohol_data.head(2)
```

```python
#select first 2 rows and 2 columns
alcohol_data.iloc[:2, :2]
```

```python
#select 2 columns
alcohol_data[['Year', 'Country']].head()
```

```python
#select random number of rows, fractions using sample
alcohol_data.sample(6)
```

```python
#select fraction of random rows using sample and frac parameter
alcohol_data.sample(frac=0.05)
```

```python
#check for missing values
alcohol_data.isnull().sum()
```

```python
#drop missing values  5 rows have been removed
alcohol_data.dropna()
```

```python
#to view the duplicates
alcohol_data['WHO region'].duplicated().sum()
```

```python
#to drop the duplicates
#syntax; df.drop_duplicates(keep=first or last; this means that python should not delete the first or the last)
alcohol_data.drop_duplicates(subset='WHO region', keep='first')
```

```python
alcohol_data.shape
```

```python
#to find out alcohol consumption of a given year (we're using filtering here)

alcohol_data[alcohol_data['Year'] == 1984]
```

```python
#to view consumption details for 1987 or 1989 using the | operator instead of 'or'
alcohol_data[(alcohol_data['Year'] == 1987) | (alcohol_data['Year'] == 1989)]
```

```python
alcohol_data[(alcohol_data['Year'] == 1985) & (alcohol_data['WHO region'] == 'Americas')]
```

```python
alcohol_data[(alcohol_data['Year'] == 1986) & (alcohol_data['WHO region'] == 'Western Pacific') & (alcohol_data['Country'] == 'Viet Nam')]
```

```python
#we are using this because we are combining 'and' 'or'
alcohol_data[((alcohol_data['Year'] == 1986) | (alcohol_data['Year'] == 1989)) & (alcohol_data['WHO region'] == 'Americas')]
```

```python
#another method
alcohol_data[(alcohol_data['WHO region'] == 'Americas') & (alcohol_data['Year'].isin([1986,1989]))]
```
